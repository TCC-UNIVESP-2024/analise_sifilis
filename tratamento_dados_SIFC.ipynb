### Transformando a coluna DT_NOTIFIC no formato data

```python
df['DT_NOTIFIC'] = pd.to_datetime(df['DT_NOTIFIC'], format = '%Y%m%d')
```

### Decodificando a coluna NU_IDADE_N
De acordo com o dicion√°rio de vari√°veis: ‚ÄúA composi√ß√£o da vari√°vel obedece o seguinte crit√©rio: 1¬∫ d√≠gito = 1. Hora , 2. Dia, 3. M√™s, 4. Ano. Ex: 3009 ‚Äî nove meses, 4018 ‚Äî dezoito anos‚Äù.


```python
df.NU_IDADE_N.unique()
```

```python
df.NU_IDADE_N.value_counts()
```


**OBSERVA√á√ÉO**: Optamos em considerar valores = 2030 como 1 m√™s de vida e excluir valores = 3000.

```python
remover_idade = df.loc[(df['NU_IDADE_N'] == 3000)]
df = df.drop(remover_idade.index)
```

```python
idade = df['NU_IDADE_N']

def decode_idade(idade):
  if idade < 2030:
    return 'Menos de 1 m√™s'
  elif idade >= 2030 and idade <4000:
    return 'Entre 1 e 11 meses'
  elif idade >= 4000:
    return 'Mais de 1 ano'
  else:
    return np.NaN

df['NU_IDADE_N'] = list(map(decode_idade, idade))
```

```python
df.NU_IDADE_N.unique()
```

```python
df.NU_IDADE_N.value_counts()
```
